## Random Variates Generation

```python
import scipy.stats
import numpy as np
#Generating random variates
data = scipy.stats.norm.rvs(size=100000, loc=0, scale=1.5, random_state=123)
hist = np.histogram(data, bins=100)
#Generating distribution
hist_dist = scipy.stats.rv_histogram(hist)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("data with normal distribution")
plt.plot(data[0:1000], label='data')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("histogram")
plt.plot(hist[0], label='histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
X = np.linspace(-5.0, 5.0, 100)
plt.title("histogram and PDF from Template")
plt.hist(data, density=True, bins=100)
plt.plot(X, hist_dist.pdf(X), label='PDF')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
X = np.linspace(-5.0, 5.0, 100)
plt.title("CDF from Template")
plt.hist(data, density=True, bins=100)
plt.plot(X, hist_dist.cdf(X), label='CDF')
plt.show()
```

## Inverse Transform

```python
from scipy.interpolate import interp1d
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
ecdf = ECDF(data)
inv_cdf = interp1d(ecdf.y, ecdf.x, bounds_error=False, assume_sorted=True)
r = np.random.uniform(0, 1, 10000)
ys = inv_cdf(r)
print(ys)
```

## Original Vs Generated Data Plot

```python
# Original Data plot
plt.hist(data, 25, histtype='step', color='orange', normed=True, linewidth=1);
# Generated Data Plot
plt.hist(ys, 25, histtype='step', color='blue', normed=True, linewidth=1);
```

## Histogram and PDF of Random Variates

```python
import matplotlib.pyplot as plt
%matplotlib inline
X = np.linspace(-5.0, 5.0, 100)
plt.title("Histogram and PDF from Random Variates")
plt.hist(ys, density=True, bins=100)
plt.plot(X, hist_dist.pdf(X), label='PDF')
plt.show()
```

## CDF of Random Variates

```python
import matplotlib.pyplot as plt
%matplotlib inline
X = np.linspace(-5.0, 5.0, 100)
plt.title("CDF from Random Variates")
plt.hist(ys, density=True, bins=100)
plt.plot(X, hist_dist.cdf(X), label='CDF')
plt.show()
```
